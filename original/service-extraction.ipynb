```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jan 26 12:31:23 2016

@author: vonw
"""

%matplotlib nbagg

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot')

import pandas as pd

from numpy           import array, max, append, mean, ones
from matplotlib.mlab import find
from netCDF4         import Dataset
from datetime        import datetime, timedelta
import seaborn as sns

# Latitude and Longitude for Pullman, WA
#dlat = 46.74377
#dlon = 360. - 117.11579
# Latitude and Longitude for Sacramento, CA
#dlat = 38.50694
#dlon = 360. - 121.495
# Latitude and Longitude for Boise, ID
dlat = 43.56704
dlon = 360. - 116.24053

# List of CMIP5 models to read.
```

```python

# models = ['BNU-ESM','CNRM-CM5','CSIRO-Mk3-6-0','bcc-csm1-1','CanESM2','GFDL-ESM2G','GFDL-ESM2M','MIROC5','MIROC-ESM','MIROC-ESM-CHEM','IPSL-CM5A-LR','IPSL-CM5A-MR','IPSL-CM5B-LR','NorESM1-M']
models = ['BNU-ESM','CNRM-CM5','CSIRO-Mk3-6-0']   # for testing...

variables = ['tasmax']
full_variables = ['precipitation', 'air_temperature']
#variables = ['was', 'vas', 'uas', 'tasmax', 'tasmin', 'rhmax', 'rhmin', 'sph', 'pr']
```

```python


maca = pd.DataFrame({})  # Initialize a pandas DataFrame.
#   ....The two Hadley Centre models only have 360 days in a year, so I didn't use them here.
for model in models:
    for variable in variables:
        print('Processing: ' + model + ' ' + variable)
        fn1 = 'http://thredds-dev.nkn.uidaho.edu:8080/thredds/dodsC/agg_macav2metdata_' + variable + '_' + model + '_r1i1p1_historical_1950_2005_CONUS_daily.nc';
        fn2 = 'http://thredds-dev.nkn.uidaho.edu:8080/thredds/dodsC/agg_macav2metdata_' + variable + '_' + model + '_r1i1p1_rcp85_2006_2099_CONUS_daily.nc';

        f1  = Dataset(fn1);
        f2  = Dataset(fn2);
    
        lat = f1.variables['lat'][:];
        lon = f1.variables['lon'][:];
        # Find the lat/lon pixel that contains the desired location
        #   THIS IS JUST APPROXIMATE FOR NOW.
    
    
        ilat = max(find(lat<dlat))
        ilon = max(find(lon<dlon))
        # Create a time vector.
        days = append(f1.variables['time'][:], f2.variables['time'][:])
        time = array([timedelta(int(day))+datetime(1900,1,1) for day in days])
        # Creater a climate vector for the particular variable and model - f1 - past - f2 future.
        
        if variable == 'pr':
             variabledata = array(append(f1.variables['precipitation'][:,ilat,ilon], f2.variables['precipitation'][:,ilat,ilon]))
        if variable == 'tasmax':
             variabledata = array(append(f1.variables['air_temperature'][:,ilat,ilon], f2.variables['air_temperature'][:,ilat,ilon]))

        
        #precip = array(append(f1.variables['precipitation'][:,ilat,ilon], f2.variables['precipitation'][:,ilat,ilon]))
    
        # Creater a precip vector.
        #maxtemp = array(append(f1.variables['air_temperature'][:,ilat,ilon], f2.variables['air_temperature'][:,ilat,ilon]))
    
        # Creater a precip vector.
        #precip = array(append(f1.variables['precipitation'][:,ilat,ilon], f2.variables['precipitation'][:,ilat,ilon]))
    
    
    
        # add a new column to the DataFrame for each model.
        maca[model + variable] = pd.Series(variabledata, index=time)


        f1.close()
        f2.close()
```

```python

maca = pd.DataFrame(maca)
maca['index'] = maca.index
```

```python
%matplotlib nbagg

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

maca.plot(x='index', y='BNU-ESMtasmax');
```

```python
# Calculate anomalies.
#Phist   = maca['1970-1-1':'1999-12-31'].mean().mean()
Phist    = maca['1970-1-1':'1999-12-31'].mean()
P2030s  = maca['2030-1-1':'2039-12-31'].mean()
P2050s  = maca['2050-1-1':'2059-12-31'].mean()
anomaly = [P2030s-Phist, P2050s-Phist]
```
